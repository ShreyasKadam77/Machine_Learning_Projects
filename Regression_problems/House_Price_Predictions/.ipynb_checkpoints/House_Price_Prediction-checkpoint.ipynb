## Feature Selection
  * It can be done in many ways :
      * 1. by using correlation matrix 
      * 2. by using feature_importance_
      * 3. by using `selectfrommodel` function in sklearn.preprocessing


**1. By using correaltion matrix**

```python
# We have to check if any features are highly correlated.
def correlation(dataset, threshold):
    col_corr = set()          # set of all names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] > threshold):        # we need absolute coeffcient value
                colname = corr_matrix.columns[i]            # column name
                col_corr.add(colname)
    return col_corr
```

## Split data into training and validation set
We have seperate test data. So for model selection and hyperparameter tuning we will use validation set.

```python
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from yellowbrick.regressor import residuals_plot
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV
```

```python
X = dataset.drop('SalePrice', axis=1)
y = dataset['SalePrice']
X.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size= 0.30, random_state=25)
```

# Build a model 
We will try 4 different models to compare their scores and select one.
* Random Forest Regressor
* Lasso
* Linear Regression
* XGBRegressor

```python
model_dict = { 'Random Forest': RandomForestRegressor(),
               'Lasso': Lasso(),
               'Linear Regression': LinearRegression(),
               'XGB': XGBRegressor()
             }
```

```python
def model_scores(models, X_train, X_val, y_train, y_val):
    '''
    This function train the model on training data (X_train, y_train) and 
    returns dictionary containing model and it's respective score as key-value pair.
    '''
    np.random.seed(0)
    
    model_score = {}
    
    for name, model in models.items():
        model.fit(X_train, y_train)
        model_score[name] = model.score(X_val, y_val)
        
    return model_score
```

```python
model_scores = model_scores(model_dict, X_train, X_val, y_train, y_val)
model_scores
```

We got good accuracy (R2 score) for two models i.e. Random Forest Regressor and XGBoost Regressor. <br>
As given in problem statement we need to evaluate a model using Root Mean Squared Error (RMSE).

```python
# We do not have built-in method to calculate RMSE. so we will create a function which gives us RMSE value.

def RMSE(y_predictions, y_true_values):
    '''
    returns root mean squared error
    '''
    return np.sqrt(mean_squared_error(y_true_values, y_predictions))
```

```python
# create a function to show all scores for model

def show_scores(model):
    '''
    returns MSE, RMSE, and R2 score for a given model.
    '''
    
    X_preds = model.predict(X_train)
    y_preds = model.predict(X_val)
    
    score={"training MSE":mean_squared_error(X_preds,y_train),
           "validation MSE":mean_squared_error(y_preds,y_val),
           "training RMSE":RMSE(X_preds,y_train),
           "validation RMSE":RMSE(y_preds,y_val),
            "training R^2":r2_score(X_preds,y_train),
           "validation R^2":r2_score(y_preds,y_val) }
    
    return score
```

**Random Forest**

```python
# check scores for Random Forest

rf = RandomForestRegressor(random_state=0)
rf.fit(X_train, y_train)
show_scores(rf)
```

**Lasso**

```python
# check scores for Lasso

lasso = Lasso(random_state=0)
lasso.fit(X_train, y_train)
show_scores(lasso)
```

**Linear Regressor**

```python
# check scores for Linear Regression

lr = LinearRegression()
lr.fit(X_train, y_train)
show_scores(lr)
```

**XGBoost**

```python
# check scores for XGBoost

xgb = XGBRegressor()
xgb.fit(X_train, y_train)
show_scores(xgb)
```

We got better RMSE for Random Forest and XGBoost. <br>
Now we will try different hyperparameters on these model to select best model.

## Hyperparameter Tuning


**Random Forest**

```python
# Create a model
rf = RandomForestRegressor(random_state=0)
rf.get_params()
```

These are all the parameters for Random Forest Regressor. <br>
By reading documentation, we get to know that n_estimators and max_features are most important parameters for this model.<br>
So we will try different values of parameters and apply Randomized SearchCV.

######  using Randomized Search CV

```python
# Take random parameters for randomized search cv

# number of trees in random forest
n_estimators = [int(x) for x in np.linspace(100, 1200, num= 12)]
# maximum number of levels in tree
max_depth = [(int(x) for x in np.linspace(5, 30, num=6)), None]
# minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
# minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
```

```python
# Create the parameter grid based on the results of random search 
param_grid = {
    'n_estimators': [int(x) for x in np.linspace(100, 1200, num= 12)],
    'max_depth': [None, 3, 6, 9, 12, 15, 18],
    'min_samples_leaf': [1, 2, 5, 10],
    'min_samples_split': [2, 5, 10, 15, 100]
            }

# Instantiate the random search model
rf_randomcv = RandomizedSearchCV(estimator = rf, param_distributions= param_grid, 
                          cv = 5, n_iter= 20, n_jobs = 1, verbose = 1)
```

```python
%%time

# fit the model 
rf_randomcv.fit(X_train, y_train)
```

```python
rf_randomcv.best_params_
```

```python
show_scores(rf_randomcv)
```

```python
# Helper function for plotting feature importance
def plot_features(columns, importances, n=20):
    df = (pd.DataFrame({"features": columns,
                        "feature_importance": importances})
          .sort_values("feature_importance", ascending=False)
          .reset_index(drop=True))
    
    sns.barplot(x="feature_importance",
                y="features",
                data=df[:n],
                orient="h")
```

```python
plot_features(X_train.columns, best_model.feature_importances_)
```
